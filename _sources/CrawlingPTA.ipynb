```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
url = 'https://pta.trunojoyo.ac.id/welcome/index/'

headers = {
    'user-agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Mobile Safari/537.36'
}

temp = []
for page in range(1, 172):
  req = requests.get(url+str(page), headers=headers)
  bs = BeautifulSoup(req.text, 'html.parser')
  items = bs.findAll('li', {'data-id':'id-1'})
  for item in items:
    # temp = []
    title = item.find('a', 'title').text
    writer = item.find('span', '').text.replace('Penulis :', '')
    abstrac = item.find('p', {'align': 'justify'}).text
    temp.append([title, writer, abstrac])
print(temp)
```

```python
df = pd.DataFrame(temp, columns=['Judul', 'Penulis', 'Abstrak'])
df
```
