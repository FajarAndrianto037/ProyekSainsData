```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import PCA
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
# Load dataset abstrak skripsi
df = pd.read_csv('/content/drive/MyDrive/ProyekSainsData/tugas/uts/dataset-pta-fix.csv')
df
```

```python
# Membagi dataset menjadi data training dan testing
X_train, X_test, y_train, y_test = train_test_split(df['Abstrak'], df['Label'], test_size=0.2, random_state=42)

# Membuat vektor fitur dengan menggunakan CountVectorizer
vectorizer = CountVectorizer(stop_words='english')
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)

# Melakukan reduksi data dengan menggunakan PCA
pca = PCA(n_components=100)
X_train_pca = pca.fit_transform(X_train_vectorized.toarray())
X_test_pca = pca.transform(X_test_vectorized.toarray())

# Melakukan normalisasi data dengan menggunakan StandardScaler
scaler = StandardScaler()
X_train_pca = scaler.fit_transform(X_train_pca)
X_test_pca = scaler.transform(X_test_pca)

# Membuat model klasifikasi Gaussian Naive Bayes
nb = GaussianNB()
nb.fit(X_train_pca, y_train)

# Membuat model klasifikasi K-Nearest Neighbor
knn = KNeighborsClassifier()
knn.fit(X_train_pca, y_train)

# Melakukan prediksi terhadap data testing
y_pred_nb = nb.predict(X_test_pca)
y_pred_knn = knn.predict(X_test_pca)

# Menghitung akurasi dari model klasifikasi
accuracy_nb = accuracy_score(y_test, y_pred_nb)
accuracy_knn = accuracy_score(y_test, y_pred_knn)

print('Akurasi Naive Bayes: {:.2f}%'.format(accuracy_nb * 100))
print('Akurasi K-Nearest Neighbor: {:.2f}%'.format(accuracy_knn * 100))
```
