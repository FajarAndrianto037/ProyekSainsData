```python
import pandas as pd
import numpy as np
```

```python
# load the dataset
dataframe = pd.read_csv('/content/drive/MyDrive/data/airline-passengers.csv',  usecols=[1], engine='python')
data = dataframe.values
data = data.astype('int')
```

```python
# Membuat sample input dan output
look_back = 2
input_data, output_data = [], []
for i in range(len(data)-look_back-1):
    a = data[i:(i+look_back), 0]
    input_data.append(a)
    output_data.append(data[i + look_back, 0])

input_data = np.array(input_data)
output_data = np.array(output_data)
```

```python
# Menampilkan sample input dan output
print("Sample input:")
print(input_data[:20])
print("\nSample output:")
print(output_data[:20])

df = pd.DataFrame({'xt-2': input_data[:,0], 'xt-1': input_data[:,1], 'xt': output_data})
print(df)
df.to_csv('airlines_data.csv', index=False, columns=['xt-2', 'xt-1', 'xt'])
```
